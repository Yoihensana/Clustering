```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate a dataset with 500 points, with noise = 0.2
X, y = make_moons(n_samples=500, noise=0.3, random_state=42)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.title('Noisy Moons Dataset')
plt.show()

```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

# Generate a dataset with 1000 points, noise = 0.1
X, y = make_circles(n_samples=1000, noise=0.01, factor=0.5, random_state=42)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.title('Noisy Circles Dataset')
plt.show()
```
